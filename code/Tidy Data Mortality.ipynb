```python
import pandas as pd
import numpy as np
import os
import re


#load in data from file
comp_mort = pd.read_csv("../data/comp_mort_99-16.txt",
                       delimiter = '\t',
                       header=0,
                       names=["Notes", "County", "County_Code","Infant_Age_Groups",
                              "Infant Age Groups Code","Deaths","Population","Crude_Rate"],
                       nrows = 1925 )

#delete empty or redudant columns
del comp_mort['Notes']
del comp_mort['Infant Age Groups Code']

#splits the county column into form [county,state]
count = comp_mort.County.str.split(',')

#create list of counties and list of states
counties = np.empty(0)
states = np.empty(0)
for i in count:
    #print(i) #row
    #print(i[0]) #county
    #print(i[1]) #state
    counties = np.append(counties,i[0])
    states = np.append(states,i[1])

#delete prior county column
del comp_mort['County']

#add in tidier county and state columns
comp_mort['County'] = counties
comp_mort['State'] = states


#comp_mort
crude = comp_mort.Crude_Rate.str.split()

reliable = np.empty(0)
crude_rate = np.empty(0)
for line in crude:
    #print(len(line))
    #print(line[0])
    if len(line) == 1:
        reliable = np.append(reliable,0)
    else:
        reliable = np.append(reliable,1)
    crude_rate = np.append(crude_rate, line[0])

    
del comp_mort['Crude_Rate']
comp_mort['Crude_Rate'] = crude_rate
comp_mort['Unreliable'] = reliable
comp_mort['Crude_Rate']= pd.to_numeric(comp_mort['Crude_Rate'],errors='coerce')

#print(comp_mort.County_Code.describe())
#float. This works as it is a numberical code
#print(comp_mort.Infant_Age_Groups.describe())
#object. This is correct, as the ages are catagories based on age. An int could work here, but the catagories are distinct for now
#print(comp_mort.Deaths.describe())
#float. this works as it is a count of the # of deaths
#print(comp_mort.Population.describe())
#float. this works as it is a count of the # of people
#print(comp_mort.County.describe())
#object. correct as the counties are all named places
#print(comp_mort.State.describe())
##object. correct as the states are all named places
#print(comp_mort.Crude_Rate.describe())
#float. Correct as it is an average # per 1000
#print(comp_mort.Unreliable.describe())
#float. This works, as the column contains 0/1 values, given if the unreliable flag is True or False

"""
Mason Caplin
Final Tutorial Milestone 1 Writeup
Due 9/29/2020

As a graduate student, I will be working alone on my final project. Due to this, I will not be submitting a collaboration plan with this writeup. For my project, I would like to examine reasons for the disparity in birth mortality rates among certain demographics, namely race and income. Currently, a child can be up to three times more likely to die during birth, based one his or her ethnicity. I would like to examine if this is the result of factors within a hospital, or if there is a recognizable link between the birth death rate and environmental factors during pregnancy. This would require me to compare the mortality rates with factors such as race, income, and chemical pollution rates. My question would be: which factors affect birth death rates, and how doe these rates change given race, income, and chemical pollution.

At the moment, I am working on a dataset, generated from data provided by the Center for Disease Control and Prevention (CDC), that provides information on neonatal (<1day old) mortality rates. This information is reported based on the county and includes information such as the number of deaths, reliability factor, and population of the county. This dataset is 1925 rows, representing approximately sixty-one percent of the 3141 counties in the United States. I would like to use this information to provide a per capita rate for each county, which can than be compared to data on demographics and chemical pollution.

Obtaining and formatting the data presented some challenges. The dataset was generated as a subset of a much larger dataset, owned by the CDC, and required me to select what information is relevant and how to best group this information. The data was presented as a large text file, which contained very little explanation and was presented with lists within cells. I had to read the file into a somewhat workable, if not tidy, format and then begin tidying the data. This required I remove a large amount of information that is extraneous to the dataset. The information was largely licensing agreements and a statement by the CDC. I then had to remove extraneous or redundant information and separate the lists into their own respective columns. All told, I was able to generate a table that uses counties as rows, presents summary information for each county, and is workable for both categorical and quantitative variables.  

This data is useful because it can answer questions about the neonatal mortality rates by county. The crude rate is the number of neonatal deaths per one thousand individuals. This data, along with the deaths per county, can be used to show which states and counties have the highest death rates. By comparing this information to data about chemical pollution or demographic, one can determine which areas are likely most affected by these factors. 
"""

comp_mort


```
